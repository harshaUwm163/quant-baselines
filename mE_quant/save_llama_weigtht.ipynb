```python
import torch
```

```python
def get_llama(model):
    def skip(*args, **kwargs):
        pass
    torch.nn.init.kaiming_uniform_ = skip
    torch.nn.init.uniform_ = skip
    torch.nn.init.normal_ = skip
    from transformers import LlamaForCausalLM
    model = LlamaForCausalLM.from_pretrained(model, torch_dtype='auto')
    model.seqlen = 2048
    return model
```

```python
model = '/data/llama-3-8b'
model = get_llama(model)
```

```python
weights = model.model.layers[0].self_attn.q_proj.weight.data.half().numpy()
import pickle as pkl
with open('/workspace/llama3-8b-0-q_proj.pkl', 'wb') as f:
    pkl.dump(weights, f)
```

```python
with open('/workspace/llama3-8b-0-q_proj.pkl', 'rb') as f:
    kk = pkl.load(f)
```
